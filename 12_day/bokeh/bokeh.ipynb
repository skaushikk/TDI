```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, push_notebook
```

```python
N = 5000
```

```python
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii =np.random.random(size=N) * 1.5
colors = ["#%02x%02x%02x" % (r, g, 150) for r, g in zip((50+2*x).astype(int), (30+2*y).astype(int))]
```

```python
output_notebook()
```

```python
p = figure()
p.circle(x, y, radius=radii,fill_color=colors, fill_alpha=0.6, line_color=None)
```

```python
show(p)
```

```python
TOOLS = 'hover, crosshair, pan, wheel_zoom, box_zoom, reset, tap, save, box_select, poly_select, lasso_select'
p1=figure(tools=TOOLS)
colors1 = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip((50+2*x), (30+2*y))]
p1.scatter(x, y, radius=radii,fill_color=colors1, fill_alpha=0.6, line_color=None)
colors1 = ["#%02x%02x%02x" % (150, int(g), int(r)) for r, g in zip((50+2*x), (30+2*y))]
p2 = figure(tools=TOOLS)
p2.scatter(x, y, radius=radii,fill_color=colors1, fill_alpha=0.6, line_color=None)
```

```python
show(p1)
show(p2)
```

```python
x = np.linspace(0,2*np.pi,2000)
y = np.sin(np.radians(x))
p3 = figure(tools=TOOLS, x_axis_label = 'angle', y_axis_label='Amplitude', plot_height=400, plot_width=900, y_range=(-10,10))
r = p3.line(x, y, color= "#1234aa",line_width=1 )
```

```python
show(p3)
```

```python
def update(f, w=1, A=1, phi=0):
    if f=='sin': func = np.sin
    elif f=='cos': func = np.cos
    elif f=="tan": func = np.tan
    r.data_source.data['y' ] = A * func(w*x+ phi)
    push_notebook()
```

```python
show(p3, notebook_handle=True)
```

```python
from ipywidgets import interact
interact(update, f=['sin', 'cos', 'tan'], w=(0,100), A=(1,5), phi=[0,20,0.1])
```

```python
p3.xaxis.axis_label = 'Angle (degrees)'
```

```python
push_notebook()
```
